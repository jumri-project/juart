```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

from juart.ellipsoid_phantoms.ellipsoids import SheppLogan
```

# Create 2D SheppLogan Phantom

```python
fov = [0.2, 0.2]
matrix = [256, 256]
```

```python
sl = SheppLogan(fov=fov, matrix=matrix)
img = sl.get_object()
```

```python
plt.figure()
plt.imshow(img[..., 0], cmap="gray")
plt.title("2D Shepp-Logan Phantom")
plt.axis("off")
plt.tight_layout()
plt.show()
```

## Add coil to phantom

```python
sl.add_coil()
```

```python
plt.figure()
plt.imshow(img[..., 0], cmap="gray")
plt.scatter(
    sl.coil.coil_channel_positions[:, 0] * matrix[0] + matrix[0] // 2,
    sl.coil.coil_channel_positions[:, 1] * matrix[1] + matrix[1] // 2,
    c="red",
    s=50,
    label="Coils",
)
plt.title("2D Shepp-Logan Phantom")
plt.axis("off")
plt.legend()
plt.tight_layout()
plt.show()
```
